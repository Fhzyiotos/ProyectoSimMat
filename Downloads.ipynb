## Introducción  
  
  
Las ecuaciones diferenciales son una poderosa herramienta que nos permite modelar situaciones físicas de relativa complejidad, resultando así en diversas aplicaciones útiles en industrias desde la electrónica hasta la automotriz.
Una de las mas conocidas aplicaciones es el oscilador armónico, estudiado en clase, y de manera mas específica el sistema resorte masa amortiguador. Como tema de este proyecto se planteará una situación del mundo real y se utilizará un método matemático para obtener la solución a los problemas encontrados, para en seguida implementar esta solución a un código de python que permita explotar el potencial de nuestras herramientas de trabajo y así facilitar y eficientar el trabajo.

## Contextualización  
  
El sistema de resorte masa amortiguador consta como indica el nombre de estos 3 elementos, cada cual aportando una función específica. El propósito principal de este tipo de sistema es reducir la transmisión de las vibraciones y los ruidos portados por las estucturas.

### Resorte  
  
El resorte cuenta con una constante de resorte, una longitud en reposo y una constante de atenuación. Si el resorte se encuentra en reposo eso implica que el sistema está en su posición original. Si una masa se mueve, el resorte acumula energía potencial.  
  
### Amortiguador
  
El amortiguador asegura que el sistema no continúe oscilando después de la deformación y que regrese a su posición de reposo. 



## Aplicación del problema  
  
Este tipo de sistema se encuentra presente en los amortiguadores de los coches, en robótica, industria en general. Como se explicó en el apartado anterior, el objetivo principal de un sistema de este tipo es la disipación de las vibraciones. Esta capacidad se vuelve muy relevante en el contexto de la ingenieria civil pues dentro del espectro de trabajo de esta disciplina las vibraciones son responsables de muchas fallas como son los agrietamientos en estructura, sobreesfuerzos, aflojamiento de tornillos, quiebre de soldaduras entre muchos otros problemas.
La empresa Gerb cuenta entre sus productos con una extensa gama de aisladores de vibraciones especializados en la indsutria productiva, el transporte, la infraestructura y la construcción. Estos aisladores se aplican para evitar los problemas antes mencionados. En el contexto actual, la empresa Gerb quiere concursar para ser parte del consorcio que gane la licitación para construir uno de los tramos del Tren Maya. Su aportación se centra en implementar un sistema de aislamiento de vibraciones que proporcione estabilidad y seguridad a los tramos donde estas mismas condiciones se vean puestas en peligro por la naturaleza del terreno o de la ruta.

Para desarrollar este sistema la empresa requiere realizar pruebas y simulaciones para prever el diseño óptimo que cumpla con las características necesarias. El objetivo es minimizar el movimiento resultante en la via al pasar el tren a toda velocidad por la misma.
